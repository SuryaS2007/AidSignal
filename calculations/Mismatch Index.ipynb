### Proposed Rule-Based Logic for Severity Score (1-10)

Based on the unique values in `PooledFund` (CBPF, ERF, CERF, CHF) and the distribution of `Budget` values, the following rules are proposed to assign a `Severity_Score`. These rules are illustrative and would typically be refined with domain expertise and actual crisis data.

**General Principle**: Higher `Budget` generally implies higher severity. Specific `PooledFund` types are also weighted based on their typical application in crisis response.

1.  **If `PooledFund` is 'CERF' (Central Emergency Response Fund):** Often for large-scale, high-impact crises.
    *   If `Budget` > 750,000: `Severity_Score` = 10
    *   If `Budget` > 500,000 and <= 750,000: `Severity_Score` = 9
    *   If `Budget` <= 500,000: `Severity_Score` = 8

2.  **If `PooledFund` is 'CBPF' (Country-Based Pooled Fund):** Targeted for specific country crises, can be significant.
    *   If `Budget` > 700,000: `Severity_Score` = 9
    *   If `Budget` > 400,000 and <= 700,000: `Severity_Score` = 7
    *   If `Budget` <= 400,000: `Severity_Score` = 6

3.  **If `PooledFund` is 'ERF' (Emergency Response Fund):** For rapid response to sudden-onset emergencies.
    *   If `Budget` > 600,000: `Severity_Score` = 8
    *   If `Budget` > 300,000 and <= 600,000: `Severity_Score` = 6
    *   If `Budget` <= 300,000: `Severity_Score` = 5

4.  **If `PooledFund` is 'CHF' (Common Humanitarian Fund):** Similar to CBPF, country-specific funding.
    *   If `Budget` > 650,000: `Severity_Score` = 9
    *   If `Budget` > 350,000 and <= 650,000: `Severity_Score` = 7
    *   If `Budget` <= 350,000: `Severity_Score` = 5

**Note**: These thresholds are derived empirically from the simulated `Budget` distribution and the general nature of these pooled funds. In a real-world scenario, these would be informed by domain experts or more detailed data analysis.

**Reasoning**:
Now that the rule-based logic for assigning 'Severity_Score' has been proposed and articulated in the previous markdown block, the next step is to apply this logic to the `df` DataFrame. This involves creating a function or a series of conditions to calculate the 'Severity_Score' for each row based on the values in the 'PooledFund' and 'Budget' columns.



```python
def assign_severity_score(row):
    pooled_fund = row['PooledFund']
    budget = row['Budget']

    if pooled_fund == 'CERF':
        if budget > 750000:
            return 10
        elif budget > 500000:
            return 9
        else:
            return 8
    elif pooled_fund == 'CBPF':
        if budget > 700000:
            return 9
        elif budget > 400000:
            return 7
        else:
            return 6
    elif pooled_fund == 'ERF':
        if budget > 600000:
            return 8
        elif budget > 300000:
            return 6
        else:
            return 5
    elif pooled_fund == 'CHF':
        if budget > 650000:
            return 9
        elif budget > 350000:
            return 7
        else:
            return 5
    return np.nan  # Default for any unhandled cases or if data is missing

# Apply the function to create the 'Severity_Score' column
df['Severity_Score'] = df.apply(assign_severity_score, axis=1)

# Display the DataFrame with the new 'Severity_Score' column
display(df.head())

# Display descriptive statistics for the new 'Severity_Score' column
print("\nDescriptive statistics for 'Severity_Score':")
display(df['Severity_Score'].describe())
```

The `Targeted` column will serve as a secondary factor, modifying the score based on the proportion of people targeted relative to those in need. We assume that a higher number of people `In Need` indicates higher severity, and a higher proportion of `Targeted` individuals (relative to `In Need`) implies a more significant and perhaps better-resourced, but still severe, crisis.

**Logic Steps:**

1.  **Handle Missing `In Need` values:**
    *   If `In Need` is `NaN`, `Severity_Score` = 1 (lowest severity, as scale is unknown).

2.  **Base Score from `In Need` (ranges from 1 to 9):**
    *   If `In Need` <= 1,000,000: `base_score` = 2
    *   If 1,000,000 < `In Need` <= 3,000,000: `base_score` = 4
    *   If 3,000,000 < `In Need` <= 6,000,000: `base_score` = 6
    *   If 6,000,000 < `In Need` <= 15,000,000: `base_score` = 8
    *   If `In Need` > 15,000,000: `base_score` = 9

3.  **Adjustment from `Targeted` (modifies `base_score` by -1, 0, or +1):**
    *   Calculate `Targeted_Ratio = Targeted / In Need`. Handle cases where `In Need` is 0 or `Targeted` is `NaN` by not applying the adjustment.
    *   If `Targeted_Ratio` >= 0.75 (high proportion targeted): `base_score` += 1
        *   *(Interpretation: A very high targeting percentage for a significant `In Need` population suggests a major crisis with considerable response efforts, elevating its severity perception.)*
    *   If 0.25 <= `Targeted_Ratio` < 0.75: No change to `base_score`.
    *   If `Targeted_Ratio` < 0.25 (low proportion targeted): `base_score` -= 1
        *   *(Interpretation: A low targeting percentage might suggest a less urgent, less prioritized crisis, or one where needs are overwhelming capacity, thus reducing its perceived severity relative to others.)*

4.  **Final `Severity_Score`:**
    *   The final `Severity_Score` will be the adjusted `base_score`, capped between a minimum of 1 and a maximum of 10.

**Note**: These thresholds and interpretations are illustrative and based on the statistical distribution of the provided data. In a real-world scenario, these rules would be refined through domain expertise and more nuanced analysis of humanitarian indicators.

**Reasoning**:
Now that the rule-based logic for assigning 'Severity_Score' based on 'In Need' and 'Targeted' columns has been defined, I will implement this logic as a Python function and apply it to the `df_hno_2026` DataFrame. Afterwards, I will display the first few rows of the DataFrame and descriptive statistics for the new 'Severity_Score' column to confirm the application of the logic.



```python
import numpy as np

def assign_hno_severity_score_v2(row):
    in_need = row['In Need']
    targeted = row['Targeted']

    # 1. Handle Missing In Need values
    if pd.isna(in_need):
        return 1

    # 2. Base Score from In Need
    if in_need <= 1000000:
        base_score = 2
    elif in_need <= 3000000:
        base_score = 4
    elif in_need <= 6000000:
        base_score = 6
    elif in_need <= 15000000:
        base_score = 8
    else:
        base_score = 9

    # 3. Adjustment from Targeted
    # Only apply adjustment if both In Need and Targeted are valid numbers and In Need is not zero
    if not pd.isna(targeted) and in_need > 0:
        targeted_ratio = targeted / in_need
        if targeted_ratio >= 0.75:
            base_score += 1
        elif targeted_ratio < 0.25:
            base_score -= 1

    # 4. Final Severity_Score, capped between 1 and 10
    return max(1, min(10, base_score))

# Apply the function to create the 'Severity_Score' column in df_hno_2026
df_hno_2026['Severity_Score'] = df_hno_2026.apply(assign_hno_severity_score_v2, axis=1)

# Display the DataFrame with the new 'Severity_Score' column
print("df_hno_2026 with new 'Severity_Score' column (based on In Need and Targeted):")
display(df_hno_2026.head())

# Display descriptive statistics for the new 'Severity_Score' column
print("\nDescriptive statistics for 'Severity_Score' in df_hno_2026:")
display(df_hno_2026['Severity_Score'].describe())
```

```python
import numpy as np

# Calculate Gap: requirements - funding
df_combined['Gap'] = df_combined['requirements'] - df_combined['funding']

# Calculate FPP: funding / requirements
# Handle cases where requirements is 0 to avoid division by zero
df_combined['FPP'] = np.where(
    df_combined['requirements'] > 0,
    df_combined['funding'] / df_combined['requirements'],
    np.nan  # Set to NaN if requirements is 0
)

# Calculate PFPP: cbpf_allocation / funding
# Handle cases where funding is 0 to avoid division by zero
df_combined['PFPP'] = np.where(
    df_combined['funding'] > 0,
    df_combined['cbpf_allocation'] / df_combined['funding'],
    np.nan  # Set to NaN if funding is 0
)

# Display the DataFrame with the new intermediate variables
print("First 5 rows of df_combined with Gap, FPP, and PFPP:")
display(df_combined.head())

print("\nDescriptive statistics for Gap, FPP, and PFPP:")
display(df_combined[['Gap', 'FPP', 'PFPP']].describe())
```

**Reasoning**:
To ensure that each component contributes proportionally to the 'Mismatch' index, I will first normalize the 'Severity_Score', 'Gap', 'FPP', and 'PFPP' columns to a common scale (0-1). Then, I will compute the inverse of 'FPP' and 'PFPP' to represent the funding and CBPF allocation shortfalls, respectively, as higher shortfalls should contribute to higher mismatch.



```python
from sklearn.preprocessing import MinMaxScaler

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Define columns to normalize (excluding temporary NaNs for FPP and PFPP if not applicable)
columns_to_normalize = ['Severity_Score', 'Gap', 'FPP', 'PFPP']

# Before scaling, fill NaN values in 'Gap', 'FPP', and 'PFPP' with 0 or a sensible replacement
# For 'Gap', 0 means no gap, which is lowest severity. For FPP/PFPP, 0 might mean no funding or allocation
# For normalization, it's safer to fill NaNs with a value that reflects their meaning in the context.
# For FPP/PFPP, if requirements/funding were 0, it indicates an extreme situation, which should not become 0 after filling NaN.
# For now, let's fill NaNs with 0 for Gap, FPP, and PFPP as a placeholder for normalization.
# A more robust approach might consider domain-specific imputation or exclude rows with NaNs if the data is substantial.

df_combined_normalized = df_combined.copy()

# Handle potential infinite values resulting from division by zero, before filling NaN for scaling.
# This might have been indirectly handled by np.where setting to NaN, but good to double check.
df_combined_normalized.replace([np.inf, -np.inf], np.nan, inplace=True)

# Fill NaN values for normalization. Choosing 0 as a neutral baseline. Adjust if domain expertise suggests otherwise.
df_combined_normalized['Gap'] = df_combined_normalized['Gap'].fillna(0)
df_combined_normalized['FPP'] = df_combined_normalized['FPP'].fillna(0) # If no requirements or funding, assume 0 FPP
df_combined_normalized['PFPP'] = df_combined_normalized['PFPP'].fillna(0) # If no funding or CBPF allocation, assume 0 PFPP

# Normalize the selected columns to a 0-1 scale
df_combined_normalized[columns_to_normalize] = scaler.fit_transform(df_combined_normalized[columns_to_normalize])

# Calculate inverse ratios for FPP and PFPP
# A high FPP means low mismatch, so we want (1 - FPP) to indicate mismatch
df_combined_normalized['InvFPP'] = 1 - df_combined_normalized['FPP']
df_combined_normalized['InvPFPP'] = 1 - df_combined_normalized['PFPP']

print("First 5 rows of df_combined_normalized with normalized metrics and inverse ratios:")
display(df_combined_normalized.head())

print("\nDescriptive statistics for normalized Severity_Score, Gap, InvFPP, and InvPFPP:")
display(df_combined_normalized[['Severity_Score', 'Gap', 'InvFPP', 'InvPFPP']].describe())
```
